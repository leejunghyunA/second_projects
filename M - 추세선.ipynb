```python
from lightgbm import LGBMClassifier

lgbm_clf = LGBMClassifier()


params = {'n_estimators': [200, 300, 500],
          'num_leaves': [32, 64 ],
          'max_depth':[128, 160],
          'min_child_samples':[60, 100],
          'subsample':[0.8, 1]}


for i in range(4):    
    X_train, X_test, y_train, y_test = train_test_split(x, y[i],
                                                        test_size = 0.2,
                                                        random_state = 0)

    grid_clf = GridSearchCV(lgbm_clf, param_grid = params, scoring = 'accuracy', cv = 5)
    
    grid_clf.fit(X_train, y_train, 
                 early_stopping_rounds = 30,
                 eval_metric = 'auc',
                 eval_set = [(X_train, y_train),(X_test, y_test)])
    
    print('************* Monetary 추세선 추가 ****************')
    print(f'y = y{i+1}')
    print(f'최적 하이퍼 파라미터: {grid_clf.best_params_}, 최적 평균 정확도: {grid_clf.best_score_}')
    print('\n')
```
