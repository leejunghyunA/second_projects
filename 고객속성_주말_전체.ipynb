```python
idx_유지 = cust_유지.dropna(axis = 0).index
idx_유지
```

```python
purchase_df = pd.read_csv('./purchase.csv', index_col=0) 
```

```python
import datetime

purchase_df['구매일자'] = purchase_df.구매일자.apply(lambda x: datetime.datetime.strptime(str(x), '%Y%m%d'))
```

```python
purchase_df = purchase_df[purchase_df['고객번호'].isin(idx_유지)]
purchase_df
```

```python
from datetime import datetime
  
def weekdayorweekend(d):
    x = d.weekday()
    day = ''
    
    if x < 5:
        day= '평일'
    else:  
        day= '주말'
        
    return day

purchase_df['평일/주말'] = purchase_df.구매일자.apply(lambda x: weekdayorweekend(x))
```
