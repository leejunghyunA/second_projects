# 전체 증가변동폭

```python
columns = []
for i in range(1, 6):
    column = str(i) + '_' + str(i+1) + ' 대비 ' + str(i+1) + '_' + str(i+2) + ' 변동폭'
    columns.append(column)

total_cust = pd.DataFrame(q7_q6_df, columns = columns)
total_cust
```

```python
all_cust = pd.DataFrame(index = q7_q6_df.index)
all_cust['1분기 총 매출'] = weight_plus_q1.sum(axis=1)
all_cust['2분기 총 매출'] = weight_plus_q2.sum(axis=1)
all_cust['3분기 총 매출'] = weight_plus_q3.sum(axis=1)
all_cust['4분기 총 매출'] = weight_plus_q4.sum(axis=1)
all_cust['5분기 총 매출'] = weight_plus_q5.sum(axis=1)
all_cust['6분기 총 매출'] = weight_plus_q6.sum(axis=1)
all_cust['7분기 총 매출'] = weight_plus_q7.sum(axis=1)
all_cust['8분기 총 매출'] = weight_plus_q8.sum(axis=1)

all_cust
```

# !함수!: 전체 매출 증감폭 구하는 사용자 함수 생성

```python
plus_minus_columns = []
for i in range(1, 8):
    column = str(i) + '분기 대비' + str(i+1) + '분기 증감여부'
    plus_minus_columns.append(column)
    
total_cust = pd.DataFrame(index = q7_q6_df.index, columns = plus_minus_columns)

for ind, name  in enumerate(total_cust.columns):
    before_name = all_cust.columns[ind]
    after_name = all_cust.columns[ind+1]
    # 평균 증감율 = (해당분기-전분기)/전분기
    before = all_cust[before_name] #전분기
    after = all_cust[after_name] #해당분기
    mean_ratio = (after.sum()-before.sum())/before.sum()

    # 기준금액 (전분기 매출*(1+증감율))
    mean_price = before*(1+mean_ratio)
    # 컬럼 추가
    stand_name = after_name + '비교 전분기_기준금액'
    all_cust[stand_name] = mean_price
    # 비교할 데이터만 추출
    new_cust = all_cust[[after_name, stand_name]]
 # 1. 증가할 고객 (해당분기금액 > 기준금액)
    new_plus_cust = new_cust[new_cust[after_name] > new_cust[stand_name]]
    
    # 증증가 고객 or 보통증가 고객 분류(4, 3)
    # 증가고객의 기존원본 데이터 총 매출
    plus_before = before.loc[new_plus_cust.index]
    plus_after = after.loc[new_plus_cust.index]
    
    # 증가고객의 평균 증가율
    mean_double_ratio = (plus_after.sum() - plus_before.sum())/ plus_before.sum()
        # 증가고객의 기준 금액이름
    new_plus_name = '증가고객기준_' + after_name
    # 증가고객의 기준 금액 데이터프레임에 추가
    new_plus_cust[new_plus_name] = plus_before * (1+mean_double_ratio)
    
#--나중에 대입할 증가고객 대상 -----
    # 증증가 고객
    double_plus = new_plus_cust[new_plus_cust[after_name] > new_plus_cust[new_plus_name]] # 증가기준금액보다 큰 경우
    # 보통증가 고객
    stand_plus = new_plus_cust[new_plus_cust[after_name] <= new_plus_cust[new_plus_name]] # 증가기준금액보다 작거나 같은 경우
    
#2. 감소고객 -- 
    new_minus_cust = new_cust[new_cust[after_name] < new_cust[stand_name]]
    
    # 짱감소 or 보통감소 고객 분류(1, 2)
    # 감소고객의 기존원본 데이터 총 매출
    minus_before = before.loc[new_minus_cust.index]
    minus_after = after.loc[new_minus_cust.index]
    
    # 감소고객의 평균 감소율
    minus_mean_ratio = (minus_after.sum() - minus_before.sum())/minus_before.sum()
        # 감소고객의 기준 금액이름
    new_minus_name = '감소고객기준_' + after_name
    # 감소고객의 기준 금액 추가
    new_minus_cust[new_minus_name] = minus_before * (1+minus_mean_ratio)

#--나중에 대입할 감소고객 대상--
    # 짱감소고객
    double_minus = new_minus_cust[new_minus_cust[after_name] < new_minus_cust[new_minus_name]]
    # 보통 감소고객
    stand_minus = new_minus_cust[new_minus_cust[after_name] >= new_minus_cust[new_minus_name]]

# --대입할 변화x 고객대상--
    new_same_cust = new_cust[new_cust[after_name] == new_cust[stand_name]]
    #--total_cust에 대입
    total_column_name = name
    # 증증가면 4
    total_cust.loc[double_plus.index, total_column_name] = 4
    # 보통증가면 3
    total_cust.loc[stand_plus.index, total_column_name] = 3
    # 보통감소면 2
    total_cust.loc[stand_minus.index, total_column_name] = 2
    # 짱감소면 1
    total_cust.loc[double_minus.index, total_column_name] = 1
    # 같으면 0
    total_cust.loc[new_same_cust.index, total_column_name] = 0
    
total_cust
```

# 카테고리별 증감폭 구하기

```python
columns = []
for i in range(1, 7):
    column = str(i) + '_' + str(i+1) + '증감폭 대비' + str(i+1) + '_' + str(i+2) + '변동폭'
    columns.append(column)
    
category_cust = pd.DataFrame(index = q7_q6_df.index, columns = columns)
category_cust
```

## !함수! : 카테고리별 변동폭 구하는 함수 생성 

```python
columns_li = q2_q1_df.columns

# 18개의 컬럼리스트와, 해당 컬럼 인덱스 번호 입력시 해당 카테고리의 전체 변동폭 출력
def category_df(columns_, name_idx):
    # 컬럼이름
    i = columns_li[name_idx]
    category_cust[category_cust.columns[0]] = q3_q2_df[i] - q2_q1_df[i] # 1_2증감폭 대비2_3변동폭
    category_cust[category_cust.columns[1]] = q4_q3_df[i] - q3_q2_df[i] # 2_3증감폭 대비3_4변동폭
    category_cust[category_cust.columns[2]] = q5_q4_df[i] - q4_q3_df[i] # 3_4증감폭 대비4_5변동폭
    category_cust[category_cust.columns[3]] = q6_q5_df[i] - q5_q4_df[i] # 4_5증감폭 대비5_6변동폭
    category_cust[category_cust.columns[4]] = q7_q6_df[i] - q6_q5_df[i] # 5_6증감폭 대비6_7변동폭
    category_cust[category_cust.columns[5]] = q8_q7_df[i] - q7_q6_df[i] # 6_7증감폭 대비7_8변동폭

    category_cust_ = category_cust.copy()
    return category_cust_, i
```
