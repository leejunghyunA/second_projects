```python
import pandas as pd

purchase = pd.read_csv('C:/cakd7/2차 프로젝트/purchase_유지.csv', index_col = 0)
purchase
```

## 카테고리별/금액 타입별 DESCRIBE

```python
pd.options.display.float_format = '{:.1f}'.format
pd.set_option('display.max_rows', 1000)
import matplotlib.pyplot as plt


bbcls = purchase['BBCL'].unique()
price_types = ['고가','중가','저가']

for bbcl in bbcls:
    for price_type in price_types:
        print(f'{bbcl}_{price_type}')
        # 해당 대대분류, 금액타입 데이터프레임 생성
        df = purchase[(purchase['BBCL'] == bbcl)&(purchase['PRICE_TYPE'] == price_type)]
        
        # 요약 출력
        df_describe = df.describe()
        display(df_describe)
        
        # 고객별 구매금액 산점도 시각화
        plt.scatter(df['CUSTNO'], df['PURAMT'])
        plt.show()
        
        # 이상치 계산
#         Q25 = df_describe.loc['25%', 'PURAMT']
#         Q75 = df_describe.loc['75%', 'PURAMT']
#         IQR = Q75 - Q25
#         highest_whisker = Q75 + 1.5 * IQR
#         display(df[df['PURAMT'] >= highest_whisker])
        
        # 상위 100개 출력
        display(df.sort_values('PURAMT', ascending = False).iloc[:100,:])
```
