## df1 상품별 구매건수 

```python
pivot = pd.pivot_table(df, index='CUSTNO', columns='ASSO_SCL', values='RNO', aggfunc='count')
pivot
```

### 상품별 단가 가중치 곱해주기

```python
col_list =list(pivot.columns)

for col in col_list:
    for cat in list(final_cat['ASSO_SCL']):
        if col == cat:
            index = int(final_cat[final_cat['ASSO_SCL']==cat].index[0])
            pivot[col] = pivot[col] * final_cat.loc[index,'AVGVS']
            break
        else:
            pass
            
        
pivot
```
