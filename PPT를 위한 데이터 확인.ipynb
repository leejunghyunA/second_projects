```python
def quarter(x):
    month = int(str(x)[4:6])
    
    if month in [1,2,3]:
        q = 1
    elif month in [4,5,6]:
        q = 2
    elif month in [7,8,9]:
        q = 3
    elif month in [10,11,12]:
        q = 4
    return q
        
ex['구매일자'].apply(quarter)
```

```python
purchase['구매분기'] = purchase['구매일자'].apply(quarter)
purchase.head()
```

```python
purchase_pivot = pd.pivot_table(purchase, 
                             index = '고객번호',
                             columns = ['구매연도', '구매분기'],
                             values = '구매금액',
                             aggfunc = 'sum')
purchase_pivot
```

```python
purchase_2014 = purchase_pivot.sum(axis = 0)[2014]
purchase_2014
```

```python
purchase_2015 = purchase_pivot.sum(axis = 0)[2015]
purchase_2015
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (10,6))
plt.plot(purchase_2014, color = 'salmon', linewidth = 4, label = '2014년 분기별 총 구매금액')
plt.plot(purchase_2015, color = 'red', linewidth = 4, label = '2014년 분기별 총 구매금액')
plt.xticks(list(range(1,5)))
plt.legend()
plt.show()
```

```python
purchase_m_pivot = pd.pivot_table(purchase, 
                             index = '고객번호',
                             columns = ['구매연도', '구매월'],
                             values = '구매금액',
                             aggfunc = 'sum')
purchase_m_pivot
```

```python
purchase_m_pivot_sum = purchase_m_pivot.sum(axis = 0)
purchase_m_2014 = purchase_m_pivot_sum[2014]
purchase_m_2015 = purchase_m_pivot_sum[2015]
```
