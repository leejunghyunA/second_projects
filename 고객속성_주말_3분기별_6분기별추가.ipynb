```python
import pandas as pd
from datetime import datetime
  
def weekdayorweekend_friday(d):
    date_d = datetime.strptime(str(d), '%Y%m%d')
    x = date_d.weekday()
    day = ''
    
    if x < 4:
        day= '평일'
    else:  
        day= '금토일'
        
    return day
```

```python
purchase_Q1_Q6['평일/금토일'] = purchase_Q1_Q6['PURDATE'].apply(lambda x: weekdayorweekend_friday(x))
purchase_Q2_Q7['평일/금토일'] = purchase_Q2_Q7['PURDATE'].apply(lambda x: weekdayorweekend_friday(x))
purchase_Q3_Q8['평일/금토일'] = purchase_Q3_Q8['PURDATE'].apply(lambda x: weekdayorweekend_friday(x))
```

```python
week_cust_Q1_Q6_friday = pd.pivot_table(purchase_Q1_Q6, 
                                index= 'CUSTNO',
                                columns= ['평일/금토일'],
                                values= 'PURDATE',
                                aggfunc= 'count')

week_cust_Q2_Q7_friday = pd.pivot_table(purchase_Q2_Q7, 
                                index= 'CUSTNO' ,
                                columns= ['평일/금토일'],
                                values= 'PURDATE',
                                aggfunc= 'count')

week_cust_Q3_Q8_friday = pd.pivot_table(purchase_Q3_Q8, 
                                index= 'CUSTNO' ,
                                columns= ['평일/금토일'],
                                values= 'PURDATE',
                                aggfunc= 'count')
```

```python
week_cust_df_li = [week_cust_Q1_Q6_friday, week_cust_Q2_Q7_friday, week_cust_Q3_Q8_friday]

for i in range(3):
    df = week_cust_df_li[i]
    df['금토일 비율'] = df['금토일']/(df['금토일'] + df['평일'])
    display(df)
    df.to_csv(f'C:/Users/jeoun/Documents/GitHub/study/study/2차 플젝/second_projects/생성데이터/주말비율/금토일비율_q{i+1}_q{i+6}.csv')
```
