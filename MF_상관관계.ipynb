```python
# D-DAY 사용자 함수 만들기
from datetime import datetime

def d_day(date):
    if date < datetime.strptime('19000331','%Y%m%d'):
        return (datetime.strptime('19000401','%Y%m%d') - date).days
    elif date < datetime.strptime('19000701','%Y%m%d'):
        return (datetime.strptime('19000701','%Y%m%d') - date).days
    elif date < datetime.strptime('19001001','%Y%m%d'):
        return (datetime.strptime('19001001','%Y%m%d') - date).days
    elif date < datetime.strptime('19010101','%Y%m%d'):
        return (datetime.strptime('19010101','%Y%m%d') - date).days
```
