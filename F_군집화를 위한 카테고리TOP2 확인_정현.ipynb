```python
df_li = [year2014_q1, year2014_q2, year2014_q3, year2014_q4, year2015_q1, year2015_q2, year2015_q3, year2015_q4]

# 빈 데이터 프레임
total_df = pd.DataFrame(index = year2014_q1.index)

for i in range(8):
    quater = i+1
    df = df_li[i]
    df = df.drop('purchase_sum', axis=1)
    
    # 고객 번호별로 중요 상품군 추가됨
    for cust in df.index:
        # 고객 번호 별로 담겨질 딕셔너리
        dic_1 = {} # top1
        dic_2 = {} # top2
        
        rank_top2 = df.loc[cust].sort_values(ascending = False)[:2] 
        
        # 딕셔너리로 차례대로 담기 (해당 카테고리 : 구매금액)
        dic_1[f'{rank_top2.index[0]}'] = rank_top2[0]
        dic_2[f'{rank_top2.index[1]}'] = rank_top2[1]
        
        # total_df에 추가
        column_top1 = f'{quater}분기 중요 카테고리 top_1'
        column_top2 = f'{quater}분기 중요 카테고리 top_2'
        
        total_df.loc[cust, column_top1] = [dic_1]
        total_df.loc[cust, column_top2] = [dic_2]
total_df
```

# 분기별로 카테고리 횟수 보기

```python
cate_df = pd.DataFrame(columns = total_df.columns, index = year2014_q1.columns[:-1])

for i in total_df.columns:
    df = total_df[[i]]
#     list(df.loc[1][0].keys())[0]
#     list(df.loc[19336][0][0].keys())[0]
    # 분기별 카테고리 초기화
    category = {}
    for cust in total_df.index:
        if cust == 1:
            name = total_df.loc[cust, i].keys()
            cat_name = list(name)[0]
            
        else:
            name = df.loc[cust][0][0].keys()
            cat_name = list(name)[0]
        
        # category 안에 있으면 +1, 없었으면 1
        if cat_name in category.keys():
            category[cat_name] += 1
        else:
            category[cat_name] = 1
    
    print(f'######## {i} ########')
    print(len(category))

    print(sorted(category.items()))
    print()
    
    # 데이터 프레임에 추가
    for cat, val in category.items():
        cate_df.loc[cat, i] = val
display(cate_df)
```
