### 타입별 월별 평균구매액 구하기

```python
## 월별 구매 총액 구하기
month_type사치품 = pd.pivot_table(data = purchase_need_plustype[purchase_need_plustype['TYPE'] == '사치품'],
                                index = ['CUSTNO'],
                                columns = ['PURYEAR', 'PURMONTH', 'TYPE'],
                                values = ['PURAMT'],
                                aggfunc = ['sum'])
month_type일용품 = pd.pivot_table(data = purchase_need_plustype[purchase_need_plustype['TYPE'] == '일용품'],
                                index = ['CUSTNO'],
                                columns = ['PURYEAR', 'PURMONTH', 'TYPE'],
                                values = ['PURAMT'],
                                aggfunc = ['sum'])

month_type선매품 = pd.pivot_table(data = purchase_need_plustype[purchase_need_plustype['TYPE'] == '선매품'],
                                index = ['CUSTNO'],
                                columns = ['PURYEAR', 'PURMONTH', 'TYPE'],
                                values = ['PURAMT'],
                                aggfunc = ['sum'])

display(month_type사치품), display(month_type일용품), display(month_type선매품) 
```

```python
month_mean_type사치품 = pd.DataFrame(month_type사치품.mean(axis = 1), columns = ['사치품_월별 평균'])
month_mean_type일용품 = pd.DataFrame(month_type일용품.mean(axis = 1), columns = ['일용품_월별 평균'])
month_mean_type선매품 = pd.DataFrame(month_type선매품.mean(axis = 1), columns = ['선매품_월별 평균'])
# display(month_mean_type사치품), display(month_mean_type일용품), display(month_mean_type선매품) 


month_mean_type = pd.merge(month_mean_type사치품, month_mean_type일용품, left_index=True, right_index=True, how = 'outer')
month_mean_type = pd.merge(month_mean_type, month_mean_type선매품, left_index=True, right_index=True, how = 'outer')
month_mean_type
```

```python
month_mean_type[month_mean_type['선매품_월별 평균'].isnull() == False]
```
