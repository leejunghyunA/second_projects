## 주말/평일 구분 사용자함수 생성

### 주말 사용자 함수 생성

```python
import pandas as pd
from datetime import datetime
  
def weekdayorweekend(d):
    date_d = datetime.strptime(str(d), '%Y%m%d')
    x = date_d.weekday()
    day = ''
    
    if x < 5:
        day= '평일'
    else:  
        day= '주말'
        
    return day
```

```python
purchase_Q1_Q6['평일/주말'] = purchase_Q1_Q6['PURDATE'].apply(lambda x: weekdayorweekend(x))
purchase_Q2_Q7['평일/주말'] = purchase_Q2_Q7['PURDATE'].apply(lambda x: weekdayorweekend(x))
purchase_Q3_Q8['평일/주말'] = purchase_Q3_Q8['PURDATE'].apply(lambda x: weekdayorweekend(x))
```

## 6분기별로 주말비율 피벗테이블 생성

```python
week_cust_Q1_Q6 = pd.pivot_table(purchase_Q1_Q6, 
                                index= 'CUSTNO',
                                columns= ['평일/주말'],
                                values= 'PURDATE',
                                aggfunc= 'count')

week_cust_Q2_Q7 = pd.pivot_table(purchase_Q2_Q7, 
                                index= 'CUSTNO' ,
                                columns= ['평일/주말'],
                                values= 'PURDATE',
                                aggfunc= 'count')

week_cust_Q3_Q8 = pd.pivot_table(purchase_Q3_Q8, 
                                index= 'CUSTNO' ,
                                columns= ['평일/주말'],
                                values= 'PURDATE',
                                aggfunc= 'count')
```

## 저장하기

```python
week_cust_df_li = [week_cust_Q1_Q6, week_cust_Q2_Q7, week_cust_Q3_Q8]

for i in range(3):
    df = week_cust_df_li[i]
    df['주말 비율'] = df['주말']/(df['주말'] + df['평일'])
    display(df)
    df.to_csv(f'C:/cakd7/생성데이터/주말비율/주말비율_q{i+1}_q{i+6}.csv')
```
