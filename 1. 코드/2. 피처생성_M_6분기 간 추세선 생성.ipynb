# 6분기 간 추세선 생성

### 6분기 간 추세선 값 계산 후 저장

```python
# 추체선 값 적용
import numpy as np
x = [1, 2, 3, 4, 5, 6]

df_li = [m1, m2, m3, m4, m5, m6, m7, m8]

for q in range(3):
    # 추세선을 넣을 빈 데이터프레임
    trendline = pd.DataFrame(index = m1.index, columns = m1.columns)

    for i in range(len(m1.index)):
        for j in range(len(m1.columns)):
            y = []
            y1 = df_li[q].iloc[i,j]
            y2 = df_li[q+1].iloc[i,j]
            y3 = df_li[q+2].iloc[i,j]
            y4 = df_li[q+3].iloc[i,j]
            y5 = df_li[q+4].iloc[i,j]
            y6 = df_li[q+5].iloc[i,j]
            
            y.extend([y1, y2, y3, y4, y5, y6])
            # 1차 선
            fit_line = np.polyfit(x, y, 1) 
            linear = fit_line[0]
            trendline.iloc[i,j] = linear
            
            # 2차 곡선
#             fit_curve = np.polyfit(x, y, 2) # input 의미 : x축 데이터, y축 데이터, 2차원
#             curve = fit_curve[0]
#             trendline.iloc[i,j] = curve

    display(trendline)
    
    # 저장하기
    trendline.to_csv(f'C:/cakd7/second_projects/생성데이터/추세선/6분기별추세선/m_추세선q{q+1}_q{q+5}.csv')
```
