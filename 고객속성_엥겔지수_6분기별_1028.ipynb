```python
import pandas as pd


df2= pd.read_csv('purchase2_include_pt.csv')
df1 = pd.read_csv('purchase1_include_pt.csv')
```

```python
purchase = pd.concat([df1, df2], axis=0, ignore_index=True)
purchase
```

```python
purchase.drop(columns='Unnamed: 0', axis=1, inplace=True)
```

```python
null_idx = purchase[purchase['PRICE_TYPE'].isnull()].index.tolist()
```

```python
for idx in null_idx:
    if purchase.loc[idx]['ASSO'] == 'A' and purchase.loc[idx]['MCL'] == '란제리/내의':
        purchase.loc[idx,'PRICE_TYPE'] = '고가'
    elif purchase.loc[idx]['ASSO'] == 'A' and purchase.loc[idx]['MCL'] == '웨딩':
        purchase.loc[idx,'PRICE_TYPE'] = '중가'
    elif purchase.loc[idx]['ASSO'] == 'B' and purchase.loc[idx]['MCL'] in ('성인내의','란제리','스타킹','아동내의','언더웨어행사', '덧신','홈웨어'):
        purchase.loc[idx,'PRICE_TYPE'] = '저가'
    elif purchase.loc[idx]['ASSO'] == 'B' and purchase.loc[idx]['MCL'] == '란제리브랜드':
        purchase.loc[idx,'PRICE_TYPE'] = '중가'
```

```python
purchase
```

## 고객별 구매액 총합

```python
total = pd.pivot_table(data = purchase,
                         index = 'CUSTNO',
                         
                         values = 'PURAMT',
                         aggfunc = 'sum')
total
```

## 고객별 식품 총합

```python
purchase_food = purchase[purchase['BCL'] =='식품']
purchase_food
```

```python
total_food = pd.pivot_table(data = purchase_food,
                         index = 'CUSTNO',
                         
                         values = 'PURAMT',
                         aggfunc = 'sum')
total_food
```

## 엥겔지수

```python
engel = pd.DataFrame((total_food['PURAMT'] / total['PURAMT'])*100)
engel
```

```python
engel.rename(columns={'PURAMT':'ENGEL'})
```

```python
engel.to_csv('engel_1022.csv')
```
