# ![함수 생성]: 해당 분기 카테고리의 랭킹 컬럼 생성

```python
# 데이터프레임, 컬럼 입력하면 해당 컬럼의 랭킹 부여
def rank_(df, column):
    rank = []
    df_ = df[df[column]>0]
    for rank_ in range(len(df_)):
        rank_ = rank_+1
        rank.append(rank_)
    without_zero = len(rank)
    include_zero = df[df[column]==0]
    for i in range(len(include_zero)):
        rank.append(without_zero+1)
    return rank

#함수대입
def plus_rank(df, column):
    
    #큰값 순으로 정렬
    sort_value_df = df.sort_values(column, ascending = False)
    
    # 랭킹 컬럼 생성 (rank 함수 대입)
    # 컬럼이름
    column_name = column + '_rank'
    sort_value_df[column_name] = rank_(sort_value_df , column)
    
    return sort_value_df[[column_name, column]]
```

##  ![함수 생성]: 해당 분기 카테고리의 랭킹 컬럼 한번에 출력

```python
# 가격 분기 데이터 리스트
q_money_li = [weight_plus_q1, weight_plus_q2, weight_plus_q3, weight_plus_q4, weight_plus_q5, weight_plus_q6, weight_plus_q7, weight_plus_q8]

def total_look(df, column, before_q, after_q):
    
    # 1개의 컬럼
    column_name = column + '_증감여부'
    # 고객유형 파악
    for i in range(5):
        df_cust = df[df[column_name] == i]
        
# ! 1. 위에서 생성한 rank 함수 대입
        # 대입할 전분기 데이터
        before_df = q_money_li[before_q-1]
        plus_rank_qbefore = plus_rank(before_df, column)
        # 해당 데이터의 고객유형에 해당하는 데이터만 뽑아오기
        qbefore_ = plus_rank_qbefore.loc[df_cust.index]
        
        # 대입할 해당분기 데이터
        after_df = q_money_li[after_q-1]
        plus_rank_qafter = plus_rank(after_df, column)
        qafter_ = plus_rank_qafter.loc[df_cust.index]
        
# 컬럼명 변경
        qbefore_.columns = [f'{before_q}분기_{column}_rank', f'{before_q}분기_{column}_구매액']
        qafter_.columns = [f'{after_q}분기_{column}_rank', f'{after_q}분기_{column}_구매액']
        
        # 고객유형
        df_cust_value = df_cust[[column_name]]
        
# 데이터 합쳐서 보기
        total_look = pd.concat([df_cust_value, qbefore_, qafter_], axis = 1)
    
        print(f'{before_q}분기 대비 {after_q}분기 {column} 카테고리 랭킹 확인')
        
        print(f'고객유형 {i}')
        display(total_look)
        print()
```

```python
total_look(df_q1, '가구인테리어/잡화_고가', 1, 2)
```
