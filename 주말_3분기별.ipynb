```python
from datetime import datetime
  
def weekdayorweekend(d):
    date_d = datetime.strptime(str(d), '%Y%m%d')
    x = date_d.weekday()
    day = ''
    
    if x < 5:
        day= '평일'
    else:  
        day= '주말'
        
    return day

purchase['평일/주말'] = purchase['PURDATE'].apply(lambda x: weekdayorweekend(x))
```

## 3분기별로 피벗데이터 생성

```python
week_cust_Q1_Q3 = pd.pivot_table(purchase[(purchase['PURYEAR'] == 2014) & (purchase['QUARTER'].isin([1,2,3]))], 
                                index= 'CUSTNO' ,
                                columns= ['평일/주말'],
                                values= 'PURDATE',
                                aggfunc= 'count')

week_cust_Q2_Q4 = pd.pivot_table(purchase[(purchase['PURYEAR'] == 2014) & (purchase['QUARTER'].isin([2,3,4]))], 
                                index= 'CUSTNO' ,
                                columns= ['평일/주말'],
                                values= 'PURDATE',
                                aggfunc= 'count')

week_cust_Q3_Q5 = pd.pivot_table(purchase[((purchase['PURYEAR'] == 2014) & (purchase['QUARTER'].isin([3, 4])))|
                                          ((purchase['PURYEAR'] == 2015) & (purchase['QUARTER'].isin([1])))], 
                                index= 'CUSTNO' ,
                                columns= ['평일/주말'],
                                values= 'PURDATE',
                                aggfunc= 'count')

week_cust_Q4_Q6 = pd.pivot_table(purchase[((purchase['PURYEAR'] == 2014) & (purchase['QUARTER'].isin([4])))|
                                          ((purchase['PURYEAR'] == 2015) & (purchase['QUARTER'].isin([1,2])))], 
                                index= 'CUSTNO' ,
                                columns= ['평일/주말'],
                                values= 'PURDATE',
                                aggfunc= 'count')

week_cust_Q5_Q7 = pd.pivot_table(purchase[(purchase['PURYEAR'] == 2015) & (purchase['QUARTER'].isin([1,2,3]))], 
                                index= 'CUSTNO' ,
                                columns= ['평일/주말'],
                                values= 'PURDATE',
                                aggfunc= 'count')
```
